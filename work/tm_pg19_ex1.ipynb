```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(444)
```

```python
# ==============================================================
# Defines the plotting function for the rest of the notebook.
#
def graph(func,left=-25,right=25,samp1=90,samp2=110):
    x = np.linspace(left,right,samp1)
    y = func(x)
    yp = None
    xi = np.linspace(left,right,samp2)
    yi = np.interp(xi, x, y, yp)
    
    fig, ax = plt.subplots(figsize=(6,6))
    ax.grid(color='gray', linestyle=':', linewidth=0.5)
    ax.plot(x, y, 'o', xi, yi, '.')
    
    ax.axhline(y=0, color='k', linewidth=0.5)
    ax.axvline(x=0, color='k', linewidth=0.5)
    
    plt.show()
```

```python
def f(t):
    return (t**4)+(3*t**3)-(12*t**2)+t-6
graph(f)
```

```python
def g(x):
    return np.sin(x) - np.cos(x)
graph(g,left=-5,right=5)
```

```python
def theta(alpha):
    return (np.e**alpha)+(alpha*np.log(alpha))
graph(theta)
```

```python
def x(t):
    return (np.sin(t)**2) - (np.cos(t))
graph(x,left=-5,right=5)
```
