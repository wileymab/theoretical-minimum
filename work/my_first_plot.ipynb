```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(444)
```

```python
def graph(func):
    x = np.linspace(-20,20,23)
    y = func(x)
    yp = None
    xi = np.linspace(-20,20,37)
    yi = np.interp(xi, x, y, yp)

    fig, ax = plt.subplots(figsize=(10,10))
    ax.grid(color='gray', linestyle=':', linewidth=0.5)
    ax.plot(x, y, 'o', xi, yi, '.')
    
    ax.axhline(y=0, color='k', linewidth=0.5)
    ax.axvline(x=0, color='k', linewidth=0.5)
    
    plt.show()
```

```python
def myLine(x):
    return x
```

```python
graph(myLine)
```

```python
def myOtherLine(x):
    return x**2
```

```python
graph(myOtherLine)
```
